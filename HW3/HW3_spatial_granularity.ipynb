```
from pyspark.sql.functions import *
from pyspark.sql.types import DateType, TimestampType
```

```
df = spark.read.json('/mnt/dacoursedatabricksstg/dacoursedatabricksdata/busFile')
```

```
df = df.drop("direction", "poiId", "poiId2", "probability", "loc")\
               .withColumn("id", col("_id.$oid"))\
               .withColumn("calendar", col("calendar.$numberLong").cast("bigint"))\
               .withColumn("timestamp", col("timestamp.$numberLong").cast("bigint"))\
               .withColumn("congestion", col("congestion").cast("integer"))\
               .drop("_id")\
               .drop("anomaly")\
               .drop("dateTypeEnum")
df_t = df.withColumn("timestamp", col("timestamp")/1000)
df_t = df_t.withColumn("date_timestamp", col("timestamp").cast(TimestampType()))
```

```
display(df_t)
```

```
df_spatial=df_t.select("areaId1","areaId2","areaId3","gridID","busStop","justStopped")
display(df_spatial)
```

```
justStopped = df_spatial.where(col("justStopped") == True)
justStopped.count()
```

```
area8 = justStopped.where(col("areaId1") == 8)
area8.count()
```

```
areaId2_140 = area8.where(col("areaId2") == 140)
areaId2_140.count()
```

```
areaId3_563 = areaId2_140.where(col("areaId3") == 563)
areaId3_563.count()
```

```
grid102_89 = areaId3_563.where(col("gridID") == '102,89')
grid102_89.count()
```

```
busStop2165 = areaId3_563.where(col("busStop") == 2165)
busStop2165.count()
```
