```
from pyspark.sql.functions import col, unix_timestamp, from_unixtime, udf
from pyspark.sql.types import DateType, TimestampType
```

```
df = spark.read.json('/mnt/dacoursedatabricksstg/dacoursedatabricksdata/busFile')
```

```
df = df.drop("direction", "poiId", "poiId2", "probability", "loc")\
               .withColumn("id", col("_id.$oid"))\
               .withColumn("calendar", col("calendar.$numberLong").cast("bigint"))\
               .withColumn("timestamp", col("timestamp.$numberLong").cast("bigint"))\
               .withColumn("congestion", col("congestion").cast("integer"))\
               .drop("_id")\
               .drop("anomaly")\
               .drop("dateTypeEnum")
df_t = df.withColumn("timestamp", col("timestamp")/1000)
df_t = df_t.withColumn("date_timestamp", col("timestamp").cast(TimestampType()))
display(df_t)
```

```
df_journey = df_t.select("distanceCovered","lineId","journeyPatternId","dateType").withColumnRenamed("dateType","WeekendOrNot")
display(df_journey)
```

```
sum_number = df_journey.select("distanceCovered").agg({"distanceCovered":"sum"}).collect()[0]
result = sum_number["sum(distanceCovered)"]
print(result)
```

```
line27 = df_journey.where(col("lineId") == 27)
display(line27)
```

```
sum_dist_line27 = line27.select("distanceCovered").agg({"distanceCovered":"sum"}).collect()[0]
result = sum_dist_line27["sum(distanceCovered)"]
print(result)
```

```
journey271001 = df_journey.where(col("journeyPatternId") == '00271001')
display(journey271001)
```

```
sum_dist_jour = journey271001.select("distanceCovered").agg({"distanceCovered":"sum"}).collect()[0]
result = sum_dist_jour["sum(distanceCovered)"]
print(result)
```
